```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import average_precision_score

```

```python
x=t[['Age' ]]
y=t['Survived']  
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=10)
model=LogisticRegression()
model.fit(x_train,y_train)
pred=model.predict(x_test)
score=accuracy_score(y_test, pred)
 
print(model.coef_)
print(model.intercept_)
print(score )
```

```python
x=t[[ 'Age','Sex',  'Fare','Title1']]
y=t['Survived']  
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=10)
model=LogisticRegression()
model.fit(x_train,y_train)
pred=model.predict(x_test)
score=accuracy_score(y_test, pred )
 
print(model.coef_)
print(model.intercept_)
print(score)
```

When we added these features, the accuracy level of the model becomes equal to 82.68%. In the next cell, we will replace the feature "Fare" with the feature "Pclass" which is too correlated to it.

```python
x=t[[ 'Age', 'Sex', 'Title1','Pclass']]
y=t['Survived']  
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=10)
model=LogisticRegression()
model.fit(x_train,y_train)
pred=model.predict(x_test)
score=accuracy_score(y_test, pred)
score2=model.score(x_test,y_test)
print(model.coef_)
print(model.intercept_)
print(score)
```

```python
confusion_matrix=pd.crosstab(pred,y_test,rownames=['Actual'],colnames=['predicted'])
print(confusion_matrix)
```

The explanatory power of the model has developed: The accuracy is equal to 84.91% (=(106+46)/(106+46+11+46)) while the precision level is equal to 75.43% (=43/(43+14)).


```python
from sklearn.metrics import roc_curve, roc_auc_score
tpr, fpr, thresholds= roc_curve(y_test, model.predict_proba(x_test)[:,1])
plt.title('Receiver Operating Characteristic Curve')
plt.xlim([0.0,1.0])
plt.ylim([0.0,1.0])
plt.xlabel('false positive rate')
plt.ylabel('true positive rate')
plt.plot(fpr,tpr,color='blue',lw=3)
```

```python
print('The area under curve= ',roc_auc_score(y_test, model.predict_proba(x_test)[:,1]))
```

The area under curve (AUC) shows the quality of predictions of the logistic regression with different thresholds (not only the 50% level). In our case, it is equal to 88.84% which means that the model has a good predictive quality.

# Decision trees

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier  
from sklearn.model_selection import train_test_split  
from sklearn import metrics
```

```python
x=t[['Sex','Age']]
y=t['Survived']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)  
clf=tree.DecisionTreeClassifier()
modelfit=clf.fit(x_train,y_train)
y_pred = clf.predict(x_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
x=t[['Fare','Title1']]
y=t['Survived']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)  
clf=tree.DecisionTreeClassifier()
modelfit=clf.fit(x_train,y_train)
y_pred = clf.predict(x_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
